```python
# Transformemos numeros mal formateados (transformemos ',' y '.')
# Para estos efectos definimos una función y la aplicamos a las columnas que necesitamos sean modificadas.
# En primera instancia nuestra función transformará un string o str a float o decimal, luego transformaremos ese
# decimal en un int.

valid = '1234567890.'
def sanitize(data):
    return float(''.join(filter(lambda char: char in valid, data)))
```

```python
df['Population'] = df['Population'].apply(sanitize)
```

```python
df['Population'] = df['Population'].astype(int)
```

```python
b = df.loc[df['Population'] >= 200000000]
b
```
